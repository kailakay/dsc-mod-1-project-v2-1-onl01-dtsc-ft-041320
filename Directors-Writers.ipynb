```python
import pandas as pd
```

```python
dfcrew = pd.read_csv("zippedData/imdb.title.crew.csv.gz")
dfCrew = dfcrew.set_index("tconst")
dfCrew.head()
```

```python
dfCrew = dfcrew.fillna('nm')
```

# Splitting directors and writers into list of strings and getting count

```python
def splitNewCol(dfName, columnName):
    newColumn = []
    for val in dfName[columnName]:
        newColumn.append(val.split(","))
    return(newColumn)
```

```python
dfCrew['writersList'] = splitNewCol(dfCrew, 'writers')
```

```python
dfCrew['directorsList'] = splitNewCol(dfCrew, 'directors')
```

```python
dfCrew.drop(['directors', 'writers'], axis = 1, inplace = True)
```

```python
def countNewCol(dfName, columnName):
    newCount = []
    for val in dfName[columnName]:
        newCount.append(len(val))
    return(newCount)
```

```python
dfCrew['writerCount'] = countNewCol(dfCrew, 'writersList')
```

```python
dfCrew['directorCount'] = countNewCol(dfCrew, 'directorsList')
```
